```python
import requests
import pandas as pd
import os

def download_image(url, folder, file_name):
    response = requests.get(url)
    if response.status_code == 200:
        path = os.path.join(folder, file_name)
        with open(path, 'wb') as file:
            file.write(response.content)
        print(f"Downloaded {file_name} to {folder}")
    else:
        print(f"Failed to download {url}")

def download_images_from_csv(csv_file, main_folder):
    df = pd.read_csv(csv_file)
    if not os.path.exists(main_folder):
        os.makedirs(main_folder)
    for index, row in df.iterrows():
        sub_folder = os.path.join(main_folder, row['Common_Name'].replace(" ", "_"))
        if not os.path.exists(sub_folder):
            os.makedirs(sub_folder)
        image_url = row['Link']
        file_name = f"{row['Scientific_Name']}_{index}.jpg"
        download_image(image_url, sub_folder, file_name)

csv_file = 'F:\mushroom_project\mushroom_images.csv'  # 替换为您的 CSV 文件路径
main_folder = 'Downloaded_Images'  # 所有类别文件夹将存放在这个大文件夹中
download_images_from_csv(csv_file, main_folder)
```
