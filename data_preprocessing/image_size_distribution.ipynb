```python
import PIL.Image as Image
import matplotlib.pyplot as plt
import numpy as np
import os


def plot_resolution(dataset_root_path):
    img_size_list = []  
    for root, dirs, files in os.walk(dataset_root_path):
        for file_1 in files:
            file_i_full_path = os.path.join(root, file_1)
            img_i = Image.open(file_i_full_path)
            img_i_size = img_i.size  
            img_size_list.append(img_i_size)
    
    print(img_size_list)  #

    width_list = [img_size_list[i][0] for i in range(len(img_size_list))]
    height_list = [img_size_list[i][1] for i in range(len(img_size_list))]

    # print(width_list)  # 640
    # print(height_list)  # 346

    plt.rcParams['font.size'] = 8
    plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号 "-" 如负号为False
    
    plt.scatter(width_list, height_list, s=1)
    plt.xlabel("Width")
    plt.ylabel("Height")
    plt.title("size")
    plt.show()
```

```python
dataset_root_path = "F:\mushroom_project\kaggle_images"

plot_resolution(dataset_root_path)
```

```python
def plot_bar(dataset_root_path):
    file_name_list = []  # 存放文件夹名
    file_num_list = []  # 存放文件夹内文件数量

    for root, dirs, files in os.walk(dataset_root_path):
        if len(dirs) != 0:
            for dir_i in dirs:
                file_name_list.append(dir_i)
                dir_path = os.path.join(root, dir_i)
                file_num_list.append(len(os.listdir(dir_path)))

    file_num_list = file_num_list[1:]  # 去掉根目录，此处假设根目录不统计文件

    # 计算均值，打印均值以供检查
    mean = np.mean(file_num_list)
    print("mean = ", mean)

    bar_positions = np.arange(len(file_name_list))

    fig, ax = plt.subplots()  # 生成画布和坐标轴对象

    ax.bar(bar_positions, file_num_list, 0.5)  # 绘制柱状图，宽度：自定义宽度，由此处的 0.5 设定

    ax.plot(bar_positions, [mean for _ in bar_positions], color="red")  # 绘制平均线

    plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体
    plt.rcParams['font.size'] = 8
    plt.rcParams['axes.unicode_minus'] = False  # 用来正确显示负号 "-" 如负号为False

    ax.set_xticks(bar_positions)
    ax.set_xticklabels(file_name_list, rotation=90)  # 设置x轴的标签，进行旋转
    ax.set_ylabel("文件数量")
    ax.set_title("文件夹内文件数量")

    plt.show()
```

```python
dataset_root_path = "F:\mushroom_project\kaggle_images"

plot_bar(dataset_root_path)
```
